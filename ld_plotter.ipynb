```python
df2 = pd.read_csv("r2windows.ld.tsv",sep='\t')
```

```python
df2['diff'] = df2['BP_B'] - df2['BP_A']
```

```python
for chr in open("chr_list.txt"):
    chr = chr.strip()
    chr_df = df2[(df2['CHR_A'] == chr) & (df2['CHR_B'] == chr)]
    plt.figure(figsize=(13, 3))
    sns.scatterplot(x="diff", y="R2",data=chr_df)
    plt.title(f"{chr} R2 vs Distance")
```
